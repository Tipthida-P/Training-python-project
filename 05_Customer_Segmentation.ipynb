```python
from pycaret.clustering import *
# load dataset
df_pycaret = pd.read_csv('/content/Mall_Customers.csv')
# setting preprocessing data
s = setup(df_pycaret, normalize = False,
                ignore_features = ['CustomerID'],
                ordinal_features = {'Gender' : ['Male', 'Female']},
                session_id = 123)
# เรียกใช้ clustering model ที่เราต้องการ
model_2 = create_model('kmeans', num_clusters = 5)
# plot graph
plot_model(model_2, 'cluster')
# save model
save_model(model_2, 'clustering_model_k5')
# เพิ่ม column predict cluster ใน df_pycaret
results_k5 = assign_model(model_2)

# ส่วนแสดงผล ด้วยกราฟ
# pairplot
plot_kws = {'scatter_kws': {'s': 10}}
sns.pairplot(results_k5, kind='reg', vars=results_k5.columns[2:5], plot_kws=plot_kws, hue="Cluster")
plt.show()
# scatterplot
sns.scatterplot(data=results_k5, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster")
results_k5[['Age','Annual Income (k$)','Spending Score (1-100)','Cluster']].groupby('Cluster').mean().plot.bar()
# countplot
ax = sns.countplot(data=results_k5, x="Cluster", hue="Gender", order = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'])
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='white', size=18)
```
