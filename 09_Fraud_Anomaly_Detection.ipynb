```python
corr = df.corr()
plt.figure(figsize=(10, 5))
seaborn.heatmap(corr)
```

```python
ranks = corr["Class"]

df_ranked = pd.DataFrame(ranks)
df_ranked = np.abs(df_ranked.sort_values(by="Class", ascending=False))
df_ranked.dropna(inplace=True)

df_ranked
```

```python
train_all = df[:150000]
train_c0 = train_all[df["Class"] == 0]
train_c1 = train_all[df["Class"] == 1]

# len(train_c1) ==> 293
df_sample = train_c0.sample(300)
df_train = train_c1.append(df_sample)
df_train = df_train.sample(frac=1)
```

```python
X_train = df_train.drop(["Time", "Class"], axis=1)
y_train = df_train["Class"]
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
```

```python
test_all = df[150000:]
X_test = test_all.drop(["Time", "Class"], axis=1)
y_test = test_all["Class"]
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
```
