## Supervised Learning

### Preparing Training Data

```python
data = [1,3,5,7,9]

# Prepare Training Data
print('=== Prepare Training Data ===')
X, y = [], []
for i, value in enumerate(data[:-1]):
  X.append(data[i])
  y.append(data[i+1])

print(f"X = {X}")
print(f"y = {y}")
print('\n')

# Convert to Numpy Array
print('=== Convert to Numpy Array ===')
X, y = np.array(X).reshape(-1,1), np.array(y).reshape(-1,1)
print(f"X = {X}")
print(f"y = {y}")
```

### Training Model

```python
# Training Model
reg = LinearRegression()
print('=== Training Model ===')
reg.fit(X,y)
print('Score = ',reg.score(X,y))
```

### Predict (Single Row)

```python
# Test Model
X_test = [[3]]

# print(f'Next of {X_test} is ',reg.predict(X_test))
predictions = reg.predict(X_test)
for i,predict in enumerate(predictions):
  print(f"Next of {X_test[i][0]} is {int(predict[0])}")
```

### Predict (Multiple Row)

```python
X_test = [[5],[7],[9]]
# print(f'Next of {X_test} is ',reg.predict(X_test))
predictions = reg.predict(X_test)
for i,predict in enumerate(predictions):
  print(f"Next of {X_test[i][0]} is {int(predict[0])}")
```

## Unsupervised Learning

### Preparing Training Data

```python
x = np.array([[1, 2], [1, 4], [1, 0],[10, 2], [10, 4], [10, 0]])

x_df = pd.DataFrame(x, columns=['x1', 'x2'])

px.scatter(x_df, x='x1', y='x2')
```

### Training Model

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(x)
```

### Predict (Clustering)

```python
predictions = kmeans.predict(x)
predictions
```

```python
x_df['group'] = predictions
# Convert Group to String
x_df['group'] = x_df['group'].astype(str)

px.scatter(x_df, x='x1', y='x2', color='group')
```
