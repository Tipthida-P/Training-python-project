Q1: เขียนโปรแกรม รับคะแนน นักเรียน แล้วทำการตัดเกรด 

*   A : คะแนน ตั้งแต่ 80 ขึ้นไป
*   B : คะแนน ระหว่าง 60-79
*   C : คะแนน น้อยกว่า 60



```python
score = int(input("input score: "))

if score >= 80:
  print("A")
elif score >= 60 and score <=79:
  print("B")
else:
  print("C")
```

Q2: เขียนโปรแกรม รับชื่อ แล้วทำการนับ ตัวอักษร สระ (a,e,i,o,u)

```python
name = input("input name: ")

count = { "a": 0,
          "e": 0}

word = name.split()
print("word", word)

for i in word[0]:
  if i=='a':
    count["a"] = count["a"]+1
  elif i=='e':
    count["e"] = count["e"]+1
print(count)
```
