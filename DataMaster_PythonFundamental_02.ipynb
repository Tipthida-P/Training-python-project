###Construct string from variable

```python
### example 2-02 ###

my_number = 37

new_str = str('Peter_{0}'.format(my_number))
print(new_str)

new_str = str('John_' + str(my_number))
print(new_str)

new_str = str('Nick_%d' % (my_number))
print(new_str)
```
